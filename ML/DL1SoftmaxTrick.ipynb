```python
import numpy as np
```

```python
S = np.array([1.0, 2.0, -1.0])
```

```python
def naive_softmax(S_input: np.ndarray) -> np.ndarray:
  return np.exp(S_input) / np.sum(np.exp(S_input))
```

```python
naive_softmax(S)
```

```python
np.sum(naive_softmax(S))
```

```python
S_extreme = np.array([450.0, 2.0, -1.0])
```

```python
naive_softmax(S_extreme)
```

```python
S_really_extreme = np.array([4500.0, 2.0, -1.0])
```

```python
naive_softmax(S_really_extreme)
```

```python
def softmax_with_trick(S_input: np.ndarray) -> np.ndarray:
  max_s = np.max(S_input)
  negative_or_null_S = S_input - max_s
  exp_S = np.exp(negative_or_null_S)
  return exp_S / np.sum(exp_S)
```

```python
softmax_with_trick(S)
```

```python
softmax_with_trick(S_extreme)
```

```python
softmax_with_trick(S_really_extreme)
```
